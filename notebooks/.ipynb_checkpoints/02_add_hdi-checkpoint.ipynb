```python
import pandas as pd
import pycountry
```

```python
# Load hdi datasets
hdi = pd.read_excel("../data/raw/hdi.xlsx")
birth_prob = pd.read_csv("../data/processed/birth_probability.csv")
```

```python
#cleaning rows with missing values
hdi = hdi[
    [
        "countryIsoCode",
        "country",
        "year",
        "value"
    ]
].copy()

hdi = hdi.rename(
    columns={
        "countryIsoCode": "Country Code",
        "country": "Country Name",
        "year": "Year",
        "value": "HDI"
    }
)
```

```python
# Filter out aggregated regions and keep valid countries only
valid_codes = {
    country.alpha_3
    for country in pycountry.countries
}
hdi= hdi[hdi["Country Code"].isin(valid_codes)]
```

```python
# Merge Birth Probability and HDI data
df = pd.merge(birth_prob,
              hdi,
              on=["Country Name", "Country Code", "Year"],
              how="inner")
```

```python
df.to_csv(
    "../data/processed/birth_probability_hdi.csv",
    index=False
)
```
