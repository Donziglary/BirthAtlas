# BirthAtlas — Birth Story Generator (v2.3)

Goal: turn the raw numeric output of `get_country_summary()` (v2.2) into a warm,
narrative English paragraph — the "story" a non-technical visitor actually reads.

Design principle: **numbers and narrative stay separated.** This notebook only
decides *how to phrase* the numbers; it never recomputes statistics itself. That
keeps a future tone change (or a second language) a one-file edit, not a rewrite
of the comparison engine.

Output shape: each story is returned as **both** a list of individual beats
(for step-by-step / reveal-style UI later) and a single joined paragraph
(for plain-text use, e.g. alt text or a simple copy).

````python
import pandas as pd
import numpy as np

# Re-load the corrected dataset and re-run the v2.2 comparison engine inline,
# so this notebook is runnable on its own without depending on a live kernel
# from 04_country_comparison.ipynb.

df = pd.read_csv("../data/processed/birth_probability_hdi.csv")

df["HDI_Rank"] = df.groupby("Year")["HDI"].rank(ascending=False, method="min")

yearly_avg = (
    df.groupby("Year")
    .agg(
        AvgBirthProbability=("BirthProbability", "mean"),
        AvgHDI=("HDI", "mean"),
        CountriesWithHDI=("HDI", lambda s: s.notna().sum()),
        TotalCountries=("Country Code", "nunique"),
    )
    .reset_index()
)


def get_country_summary(df: pd.DataFrame, yearly_avg: pd.DataFrame, country_code: str, year: int) -> dict:
    """Return comparison metrics for one country in one year. See 04_country_comparison.ipynb
    for the full explanation of every field; this is the same function, copied here so the
    Birth Story notebook stays independently runnable."""
    row = df[(df["Country Code"] == country_code) & (df["Year"] == year)]

    if row.empty:
        raise ValueError(
            f"No data found for country_code='{country_code}', year={year}. "
            "Check spelling/ISO code and confirm the year exists in the dataset."
        )

    row = row.iloc[0]
    avg_row = yearly_avg[yearly_avg["Year"] == year].iloc[0]
    has_hdi = pd.notna(row["HDI"])

    summary = {
        "CountryName": row["Country Name"],
        "CountryCode": country_code,
        "Year": year,
        "BirthProbability": round(float(row["BirthProbability"]), 3),
        "BirthRank": int(row["Rank"]),
        "EstimatedBirths": round(float(row["Births"])),
        "RelativeToGlobalAverage": round(float(row["BirthProbability"]) / avg_row["AvgBirthProbability"], 2),
        "HDI": round(float(row["HDI"]), 3) if has_hdi else None,
        "HDI_Rank": int(row["HDI_Rank"]) if has_hdi else None,
        "CountriesWithHigherHDI": None,
        "CountriesWithLowerHDI": None,
        "TotalCountriesThisYear": int(avg_row["TotalCountries"]),
    }

    if has_hdi:
        year_df = df[df["Year"] == year]
        summary["CountriesWithHigherHDI"] = int((year_df["HDI"] < row["HDI"]).sum())
        summary["CountriesWithLowerHDI"] = int((year_df["HDI"] > row["HDI"]).sum())

    return summary
````

## Step 1 — Probability-to-analogy helper

Raw percentages don't *feel* like anything to a casual reader. This helper maps a
`BirthProbability` value onto a familiar everyday-odds comparison, picked from a small
ladder of reference probabilities (coin flips, dice rolls, card draws). It picks the
**closest** rung below the actual probability so the analogy always reads as
"rarer than X" rather than overstating rarity.

Kept as its own function (not inlined into the story builder) so the analogy ladder
can be tuned independently later without touching narrative logic.

````python
# Each entry: (probability_as_percent, description). Sorted descending.
# Reference odds:
#   coin flip               = 50%
#   rolling a six           = 16.7%
#   drawing an ace          = 7.7%   (4/52)
#   rolling double sixes    = 2.8%   (1/36)
#   a royal flush hand part = 0.4%   (illustrative, simplified)
ODDS_LADDER = [
    (50.0, "about as likely as a single coin flip landing heads"),
    (16.7, "rarer than rolling a six on a single die"),
    (7.7, "rarer than drawing an ace from a shuffled deck"),
    (2.8, "rarer than rolling double sixes in one throw"),
    (1.0, "rarer than flipping five heads in a row"),
    (0.4, "rarer than rolling double sixes twice in a row"),
    (0.0, "rarer than almost any everyday coincidence you've experienced"),
]


def probability_to_analogy(probability_percent: float) -> str:
    """Return a human-scale odds analogy for a given BirthProbability (already in percent)."""
    for threshold, phrase in ODDS_LADDER:
        if probability_percent < threshold:
            continue
        return phrase
    return ODDS_LADDER[-1][1]


# quick check across a spread of values
for p in [45, 12, 5, 2, 0.8, 0.1]:
    print(f"{p}% -> {probability_to_analogy(p)}")
````

## Step 2 — Story beat builders

Four small functions, one per narrative beat. Each takes the summary dict (and
sometimes nothing else) and returns one English sentence. Splitting them up means
the reveal-style UI (planned for v2.5/v2.6) can request beats individually instead
of parsing a paragraph.

````python
def _beat_intro(summary: dict) -> str:
    return f"You were born in {summary['CountryName']} in {summary['Year']}."


def _beat_probability(summary: dict) -> str:
    pct = summary["BirthProbability"]
    rank = summary["BirthRank"]
    total = summary["TotalCountriesThisYear"]
    analogy = probability_to_analogy(pct)
    rel = summary["RelativeToGlobalAverage"]

    rel_phrase = (
        f"about {rel}x the global average for that year"
        if rel >= 1
        else f"only {rel}x the global average for that year"
    )

    return (
        f"Out of every birth on Earth that year, only {pct}% happened in {summary['CountryName']} "
        f"\u2014 {analogy}. That places {summary['CountryName']} {rank} out of {total} countries "
        f"by birth share, {rel_phrase}."
    )


def _beat_hdi(summary: dict) -> str | None:
    if summary["HDI"] is None:
        return None

    return (
        f"That same year, {summary['CountryName']} had a Human Development Index of "
        f"{summary['HDI']}, ranking {summary['HDI_Rank']} out of {summary['TotalCountriesThisYear']} "
        f"countries \u2014 ahead of {summary['CountriesWithLowerHDI']} of them, and behind "
        f"{summary['CountriesWithHigherHDI']}."
    )


def _beat_hdi_fallback(summary: dict) -> str:
    # Used only when HDI data is missing, instead of an empty/awkward gap.
    return (
        f"Human development data for {summary['CountryName']} isn't available for {summary['Year']}, "
        "but the birth odds alone already tell part of the story."
    )


def _beat_closing(summary: dict) -> str:
    return (
        "None of this was chosen \u2014 it's the birth lottery: the place and moment you arrived, "
        "decided before you ever had a say."
    )
````

## Step 3 — Public function: `generate_birth_story`

Combines the beats above and returns both forms of output.

```python
{
  "beats": [str, str, str, str],   # always 4 items, in display order
  "paragraph": str                 # all beats joined with spaces
}
```

````python
def generate_birth_story(summary: dict) -> dict:
    """Turn a get_country_summary() dict into a narrative English Birth Story.

    Returns a dict with:
      - "beats": list[str] of 4 sentences in display order (intro, probability, HDI, closing)
      - "paragraph": str, all beats joined into one paragraph

    If HDI data is missing for this country/year, beat 3 becomes a graceful fallback
    sentence instead of being silently dropped \u2014 the story always has 4 beats.
    """
    beats = [
        _beat_intro(summary),
        _beat_probability(summary),
        _beat_hdi(summary) or _beat_hdi_fallback(summary),
        _beat_closing(summary),
    ]

    return {
        "beats": beats,
        "paragraph": " ".join(beats),
    }
````

## Step 4 — Test: normal case (Iran, 2007)

````python
summary = get_country_summary(df, yearly_avg, country_code="IRN", year=2007)
story = generate_birth_story(summary)

for i, beat in enumerate(story["beats"], 1):
    print(f"{i}. {beat}\n")

print("--- Full paragraph ---")
print(story["paragraph"])
````

## Step 5 — Test: edge case (country/year missing HDI)

Reuses the same discovery approach from 04_country_comparison.ipynb to find a real
missing-HDI row in the corrected dataset, then confirms the fallback sentence reads
naturally instead of looking like an error message.

````python
missing_hdi = df[df["HDI"].isna()][["Country Name", "Country Code", "Year"]]

if len(missing_hdi) > 0:
    sample = missing_hdi.iloc[0]
    edge_summary = get_country_summary(df, yearly_avg, country_code=sample["Country Code"], year=int(sample["Year"]))
    edge_story = generate_birth_story(edge_summary)

    for i, beat in enumerate(edge_story["beats"], 1):
        print(f"{i}. {beat}\n")
else:
    print("No missing-HDI rows found \u2014 edge case does not apply to this dataset.")
````
