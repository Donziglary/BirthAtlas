# BirthAtlas — Birth Lottery Simulator (v2.4)

Goal: simulate "if you had been born completely at random somewhere on Earth, sometime
between 1993 and 2020, where and when would that most likely have been?"

This notebook reuses the engines built in v2.2 (`get_country_summary`) and v2.3
(`generate_birth_story`) without modifying either \u2014 it only adds the random-selection
logic on top, then feeds the result through the existing functions.

## Sampling design

A truly random birth is **not** uniform across countries (Monaco and China should not
have equal odds) and **not** uniform across years either (the world had far fewer births
in 1993 than in the early 2010s, so a random birth is more likely to land in a
higher-birth year).

Two-stage weighted sampling models this correctly:

1. **Pick a year**, weighted by that year's total global births \u2014 years with more
   births happening worldwide are proportionally more likely to be picked.
2. **Pick a country within that year**, weighted by `BirthProbability` \u2014 exactly the
   column the whole project already computes, so no new statistic is introduced here.

This keeps the simulator consistent with the project's existing definition of
"birth probability" instead of inventing a second, slightly different notion of
randomness.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/processed/birth_probability_hdi.csv")

df["HDI_Rank"] = df.groupby("Year")["HDI"].rank(ascending=False, method="min")

yearly_avg = (
    df.groupby("Year")
    .agg(
        AvgBirthProbability=("BirthProbability", "mean"),
        AvgHDI=("HDI", "mean"),
        CountriesWithHDI=("HDI", lambda s: s.notna().sum()),
        TotalCountries=("Country Code", "nunique"),
        TotalBirths=("Births", "sum"),
    )
    .reset_index()
)


def get_country_summary(df: pd.DataFrame, yearly_avg: pd.DataFrame, country_code: str, year: int) -> dict:
    """Same function as in 04_country_comparison.ipynb \u2014 copied here so this notebook
    stays independently runnable. See that notebook for the full explanation."""
    row = df[(df["Country Code"] == country_code) & (df["Year"] == year)]

    if row.empty:
        raise ValueError(
            f"No data found for country_code='{country_code}', year={year}. "
            "Check spelling/ISO code and confirm the year exists in the dataset."
        )

    row = row.iloc[0]
    avg_row = yearly_avg[yearly_avg["Year"] == year].iloc[0]
    has_hdi = pd.notna(row["HDI"])

    summary = {
        "CountryName": row["Country Name"],
        "CountryCode": country_code,
        "Year": year,
        "BirthProbability": round(float(row["BirthProbability"]), 3),
        "BirthRank": int(row["Rank"]),
        "EstimatedBirths": round(float(row["Births"])),
        "RelativeToGlobalAverage": round(float(row["BirthProbability"]) / avg_row["AvgBirthProbability"], 2),
        "HDI": round(float(row["HDI"]), 3) if has_hdi else None,
        "HDI_Rank": int(row["HDI_Rank"]) if has_hdi else None,
        "CountriesWithHigherHDI": None,
        "CountriesWithLowerHDI": None,
        "TotalCountriesThisYear": int(avg_row["TotalCountries"]),
    }

    if has_hdi:
        year_df = df[df["Year"] == year]
        summary["CountriesWithHigherHDI"] = int((year_df["HDI"] < row["HDI"]).sum())
        summary["CountriesWithLowerHDI"] = int((year_df["HDI"] > row["HDI"]).sum())

    return summary


ODDS_LADDER = [
    (50.0, "about as likely as a single coin flip landing heads"),
    (16.7, "rarer than rolling a six on a single die"),
    (7.7, "rarer than drawing an ace from a shuffled deck"),
    (2.8, "rarer than rolling double sixes in one throw"),
    (1.0, "rarer than flipping five heads in a row"),
    (0.4, "rarer than rolling double sixes twice in a row"),
    (0.0, "rarer than almost any everyday coincidence you've experienced"),
]


def probability_to_analogy(probability_percent: float) -> str:
    for threshold, phrase in ODDS_LADDER:
        if probability_percent < threshold:
            continue
        return phrase
    return ODDS_LADDER[-1][1]


def _beat_intro(summary: dict) -> str:
    return f"You were born in {summary['CountryName']} in {summary['Year']}."


def _beat_probability(summary: dict) -> str:
    pct = summary["BirthProbability"]
    rank = summary["BirthRank"]
    total = summary["TotalCountriesThisYear"]
    analogy = probability_to_analogy(pct)
    rel = summary["RelativeToGlobalAverage"]

    rel_phrase = (
        f"about {rel}x the global average for that year"
        if rel >= 1
        else f"only {rel}x the global average for that year"
    )

    return (
        f"Out of every birth on Earth that year, only {pct}% happened in {summary['CountryName']} "
        f"\u2014 {analogy}. That places {summary['CountryName']} {rank} out of {total} countries "
        f"by birth share, {rel_phrase}."
    )


def _beat_hdi(summary: dict):
    if summary["HDI"] is None:
        return None
    return (
        f"That same year, {summary['CountryName']} had a Human Development Index of "
        f"{summary['HDI']}, ranking {summary['HDI_Rank']} out of {summary['TotalCountriesThisYear']} "
        f"countries \u2014 ahead of {summary['CountriesWithLowerHDI']} of them, and behind "
        f"{summary['CountriesWithHigherHDI']}."
    )


def _beat_hdi_fallback(summary: dict) -> str:
    return (
        f"Human development data for {summary['CountryName']} isn't available for {summary['Year']}, "
        "but the birth odds alone already tell part of the story."
    )


def _beat_closing(summary: dict) -> str:
    return (
        "None of this was chosen \u2014 it's the birth lottery: the place and moment you arrived, "
        "decided before you ever had a say."
    )


def generate_birth_story(summary: dict) -> dict:
    beats = [
        _beat_intro(summary),
        _beat_probability(summary),
        _beat_hdi(summary) or _beat_hdi_fallback(summary),
        _beat_closing(summary),
    ]
    return {"beats": beats, "paragraph": " ".join(beats)}
```

## Step 1 — Pick a random year, weighted by total global births

Uses `yearly_avg["TotalBirths"]` (computed above) as sampling weights via
`np.random.choice`. A `random_state` parameter is threaded through every function
in this notebook so a specific simulation can be reproduced for debugging or testing.

```python
def pick_random_year(yearly_avg: pd.DataFrame, random_state: int | None = None) -> int:
    """Pick one year, with probability proportional to that year's total global births."""
    rng = np.random.default_rng(random_state)
    weights = yearly_avg["TotalBirths"] / yearly_avg["TotalBirths"].sum()
    return int(rng.choice(yearly_avg["Year"], p=weights))


# sanity check: weights must sum to ~1.0
weights_check = yearly_avg["TotalBirths"] / yearly_avg["TotalBirths"].sum()
print(f"Year weights sum to: {weights_check.sum():.6f}")
print(pick_random_year(yearly_avg, random_state=42))
```

## Step 2 — Pick a random country within that year, weighted by `BirthProbability`

Within a single year, `BirthProbability` values across countries already sum to (close
to) 100% by construction \u2014 they are each country's share of that year's global births.
Renormalizing (`/ sum()`) guards only against floating-point rounding, not a meaningful
correction.

```python
def pick_random_country(df: pd.DataFrame, year: int, random_state: int | None = None) -> str:
    """Pick one country code within a given year, weighted by BirthProbability."""
    rng = np.random.default_rng(random_state)
    year_df = df[df["Year"] == year]

    weights = year_df["BirthProbability"] / year_df["BirthProbability"].sum()
    return str(rng.choice(year_df["Country Code"], p=weights))


# sanity check on one year
sample_year = 2007
year_slice = df[df["Year"] == sample_year]
print(f"BirthProbability sums to {year_slice['BirthProbability'].sum():.2f}% in {sample_year}")
print(pick_random_country(df, sample_year, random_state=42))
```

## Step 3 — Public function: `simulate_random_birth`

Combines both sampling steps with the existing v2.2/v2.3 engines. `year` is optional:

- If omitted, both year and country are randomized (full simulation).
- If provided, only the country is randomized within that fixed year.

```python
def simulate_random_birth(
    df: pd.DataFrame,
    yearly_avg: pd.DataFrame,
    year: int | None = None,
    random_state: int | None = None,
) -> dict:
    """Simulate a random birth and return its full summary + narrative story.

    Parameters
    ----------
    year : if None, the year itself is also randomized (weighted by global births
           that year). If given, only the country is randomized within that year.
    random_state : optional seed for reproducible simulations.

    Returns
    -------
    dict with keys: "summary" (from get_country_summary) and "story" (from
    generate_birth_story).
    """
    rng_seed_year = random_state
    rng_seed_country = None if random_state is None else random_state + 1

    chosen_year = year if year is not None else pick_random_year(yearly_avg, random_state=rng_seed_year)
    chosen_country = pick_random_country(df, chosen_year, random_state=rng_seed_country)

    summary = get_country_summary(df, yearly_avg, country_code=chosen_country, year=chosen_year)
    story = generate_birth_story(summary)

    return {"summary": summary, "story": story}
```

## Step 4 — Test: single simulation

```python
result = simulate_random_birth(df, yearly_avg, random_state=7)

print(f"Simulated birth: {result['summary']['CountryName']}, {result['summary']['Year']}\n")
for beat in result["story"]["beats"]:
    print(beat)
```

## Step 5 — Test: repeated simulations (required by the roadmap)

Runs several simulations in a row with no fixed seed, confirming the function can be
called repeatedly to produce a fresh random outcome each time \u2014 exactly the
"Allow repeated simulations" requirement from the roadmap.

```python
for i in range(5):
    r = simulate_random_birth(df, yearly_avg)
    print(f"{i + 1}. {r['summary']['CountryName']} ({r['summary']['Year']}) "
          f"\u2014 BirthProbability: {r['summary']['BirthProbability']}%")
```

## Step 6 — Test: fixed year, randomized country only

```python
for i in range(3):
    r = simulate_random_birth(df, yearly_avg, year=2015)
    print(f"{i + 1}. {r['summary']['CountryName']} (year fixed at 2015) "
          f"\u2014 BirthProbability: {r['summary']['BirthProbability']}%")
```

## Step 7 — Distribution sanity check

Runs many simulations and confirms higher-`BirthProbability` countries are chosen
more often \u2014 a basic check that the weighting is actually doing something, not
behaving like uniform sampling by accident.

```python
n_runs = 2000
outcomes = [simulate_random_birth(df, yearly_avg)["summary"]["CountryName"] for _ in range(n_runs)]

top_10_simulated = pd.Series(outcomes).value_counts(normalize=True).head(10) * 100
print(f"Most frequent countries across {n_runs} simulations (% of runs):")
top_10_simulated.round(2)
```
