```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
import plotly.offline as pyo
import numpy as np
```

```python
pio.renderers.default = "notebook"
```

```python
df = pd.read_csv(
    "../data/processed/birth_probability.csv"
)
```

```python
# Color Design
BIRTH_COLORSCALE = [
    [0.00, "#07111E"],
    [0.15, "#0D2137"],
    [0.32, "#0E4D8A"],
    [0.52, "#1A7BBF"],
    [0.70, "#E8813A"],
    [0.86, "#F5A623"],
    [1.00, "#FFF5CC"],
]

BG     = "#080D16"
OCEAN  = "#0B1826"
LAND   = "#152130"
BORDER = "#253545"
TEXT   = "#CBD6E2"
DIM    = "#5D7A8E"
ACCENT = "#F5A623"
```

```python
# Figure
fig = px.choropleth(
    df,
    locations="Country Code",
    color="BirthProbability",
    hover_name="Country Name",
    hover_data={
        "BirthProbability": ":.3f",
        "Births": ":,.0f",
        "Rank": True,
        "Country Code": False,
    },
    animation_frame="Year",
    color_continuous_scale=BIRTH_COLORSCALE,
    range_color=(
        df["BirthProbability"].min(),
        df["BirthProbability"].max(),
    ),
)
```

```python
# Trace styling
fig.update_traces(
    marker_line_color=BORDER,
    marker_line_width=0.35,
)

for frame in fig.frames:
    for trace in frame.data:
        trace.marker.line.color = BORDER
        trace.marker.line.width = 0.35
```

```python
fig.show()
```

```python
# Layout
fig.update_layout(
    paper_bgcolor=BG,
    font=dict(
        family="Inter, Helvetica Neue, Arial, sans-serif",
        color=TEXT,
    ),
    title=dict(
        text="<b>BirthAtlas</b>",
        x=0.5,
        xanchor="center",
        y=0.96,
        font=dict(size=32, color="#E8EDF5"),
    ),
    annotations=[dict(
        text="Probability of being born in each country · World Bank Data · 1993–2020",
        x=0.5, y=0.915,
        xref="paper", yref="paper",
        showarrow=False,
        font=dict(size=11, color=DIM),
        xanchor="center",
    )],
    geo=dict(
        bgcolor=OCEAN,
        showocean=True, oceancolor=OCEAN,
        showland=True, landcolor=LAND,
        showlakes=False, showrivers=False,
        showcoastlines=True, coastlinecolor=BORDER, coastlinewidth=0.5,
        showframe=False,
        showcountries=True, countrycolor=BORDER, countrywidth=0.35,
        projection_type="natural earth",
    ),
    coloraxis_colorbar=dict(
        title=dict(text="Birth Share %", font=dict(size=11, color=TEXT)),
        tickfont=dict(size=10, color=DIM),
        tickformat=".2f",
        thickness=12,
        len=0.55,
        x=1.01,
        bgcolor="rgba(11,24,38,0.7)",
        outlinewidth=0,
        borderwidth=0,
    ),
    margin=dict(l=0, r=90, t=68, b=55),
    sliders=[dict(
        currentvalue=dict(
            prefix="Year: ",
            font=dict(size=13, color=TEXT),
            visible=True,
            xanchor="center",
        ),
        pad=dict(t=15, b=5),
        font=dict(color=DIM, size=10),
        bgcolor="#0F1D2B",
        activebgcolor=ACCENT,
        bordercolor=BORDER,
        tickcolor=BORDER,
        x=0.05,
        len=0.9,
    )],
    updatemenus=[dict(
        type="buttons",
        showactive=False,
        y=-0.11,
        x=0.5,
        xanchor="center",
        buttons=[
            dict(
                label="▶ Play",
                method="animate",
                args=[None, dict(frame=dict(duration=450, redraw=True), fromcurrent=True)],
            ),
            dict(
                label="⏸ Pause",
                method="animate",
                args=[[None], dict(frame=dict(duration=0, redraw=False), mode="immediate")],
            ),
        ],
        font=dict(color=TEXT, size=12),
        bgcolor="#0F1D2B",
        bordercolor=BORDER,
    )],
)
fig.show()
```
